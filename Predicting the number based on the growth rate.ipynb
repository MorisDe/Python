```python
def next_number(n,i,j):
    a=n[i]
    b=n[j]
    growth_rate=((b-a)/b)
    growth_factor=1+growth_rate
    next_number=growth_factor*b
    return next_number
```

```python
no=52
n=[5,6,7,55,74,98]
start=4
next=5
for i in range(0,no):
  k=next_number(n,start,next)
  n.append(k)
  start=start+1
  next=next+1
  
print(n)
```

```python
plt.plot(n)
plt.show()
```
